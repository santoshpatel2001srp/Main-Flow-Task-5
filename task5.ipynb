```python
# plotting a pie chart
df.sex.value_counts().plot(kind="pie", figsize=(8, 6))
plt.title("Male Female ratio")
plt.legend(["Male", "Female"]);
```

```python
# Q2.  People with which sex has the most heart disease?
pd.crosstab(df.target, df.sex)
```

```python
# Q5. Is there a correlation between exercise-induced angina and heart disease?
pd.crosstab(df.exang, df.target)
```

```python
grouped = df.groupby(['exang', 'target']).size().unstack()

# Plotting
grouped.plot(kind='bar')
plt.xlabel('Exercise Induced Angina (exang)')
plt.ylabel('Number of Patients')
plt.title('Exang vs Target')
plt.legend(title='Heart Disease (target)')
plt.show()
```

```python
# the graph shows that the people who exercize more have low rate of people with Exercise-induced Anigma i.e. the low heart disease
```

```python
# Q6. Does 'fbs' affect the rate of 'Heart Disease'?

# Group by fbs and target
grouped = df.groupby(['fbs', 'target']).size().unstack()

# Plotting
plt.figure(figsize=(10, 6))
grouped.plot(kind='bar', color=['lightgreen','salmon'])
plt.xlabel('Fasting Blood Sugar (fbs), 0 =  <120mg/dl, 1 = >120mg/dl')
plt.ylabel('Number of Patients')
plt.title('Fasting Blood Sugar vs Heart Disease')
plt.legend(title='Heart Disease (target)')
plt.show()
```

```python
# Patients with Fasting Blood Sugar < 120mg/dl:
# More patients do not have heart disease compared to those who do.
# Patients with Fasting Blood Sugar ≥ 120mg/dl:
# More patients have heart disease compared to those who do not.
```

```python
# Q7. How can one predict heart disease based on the coronary artery?
pd.crosstab(df.ca, df.target)
```

```python
sns.countplot(x='ca', data=df, hue='target');
```

```python
# Key Insights:
# ca = 0: Higher count of patients without heart disease.
# ca = 1, 2, 3: Higher count of patients with heart disease.
# ca = 4: Very few patients, but more likely to have heart disease.
```
